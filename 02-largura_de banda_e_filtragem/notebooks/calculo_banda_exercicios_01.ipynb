# Exercício 2: O Limite de -3dB

Um receptor de rádio utiliza um filtro sintonizado em f0​=210 Hz com um Fator de Qualidade Q=2.33.
image.png
 A resposta em frequência do filtro gerou a imagem acima e é dada pela função de magnitude:

$$|H(f)| = \frac{1}{\sqrt{1 + \left( \frac{f^2 - f_0^2}{f \cdot B} \right)^2}}$$

Dadas as observações:

OBS1: **O ponto de $-3$ dB** ocorre quando o termo dentro do parêntese da fórmula é igual a $\pm 1$.

OBS2: $-3$ dB corresponde a aproximadamente $50\%$ da potência máxima ($10 \log_{10}(0.5) \approx -3.01$ dB).

Calcule:

  a) A largura de banda B do dispositivo:
  $$B=f_0​/Q$$
  
  b) $f_1$ (Corte Inferior)
  
  c) $f_2$ (Corte Superior)

# Exercício 3: Filtro

Você quer sintonizar uma rádio que opera na frequência central de **$f_0 = 100 \text{ MHz}$**. No entanto, existem duas estações vizinhas "sujando" o sinal: uma em **90 MHz** e outra em **110 MHz**.

a) Qual tipo de filtro deve ser usado para isolar a estação de 100 MHz e eliminar as vizinhas de 95 e 110 MHz?

b) Calcule a Largura de Banda ($B$) necessária para que o filtro cubra exatamente o sinal central sem tocar nas vizinhas.

c) Calcule o Fator de Qualidade ($Q$) deste filtro.

**Fórmula:**  
$$Q = \frac{f_0}{B}$$
